# Create pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, log_loss, make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import MinMaxScaler, FunctionTransformer, PowerTransformer
from sklearn.multiclass import OneVsRestClassifier
import numpy as np
```

```python
X_train = X
y_train = y
```

```python
svd = TruncatedSVD(n_components = 5)
power_transf = PowerTransformer(method = 'yeo-johnson', standardize = True)
```

```python
numerical_transformer = Pipeline(steps = [
    ('power', power_transf)
])
categorical_transformer = Pipeline(steps =  [
    ('svd', svd)
])
```

```python
preprocessor = ColumnTransformer( transformers = [
    ('num', numerical_transformer, numerical_col),
    ('cat', categorical_transformer, cat_col)
])
```

```python
rf_clf = RandomForestClassifier(random_state = 0)
ovr = OneVsRestClassifier(rf_clf)
```

```python
pipeline = Pipeline(steps = [
    ('preprocessor',preprocessor),
    ('clf', ovr)
])
```

```python
labels = np.unique(y)
print(labels)
log_loss_build = make_scorer(log_loss, 
                             greater_is_better=False,
                             needs_proba=True, 
                             labels=sorted(labels))
```

Predict the training and the validation dataset to calculate the accuracy

```python
import time
from sklearn.model_selection import cross_val_score
```

# Training XGBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
```

```python
xgb = XGBClassifier(objective = 'binary:logistic')
ovr_xgb = OneVsRestClassifier(xgb)
```

```python
pipeline_xgb = Pipeline (steps =[
    ('preprocessor',preprocessor),
    ('xgb', ovr_xgb)
])
```

```python
param_grid_xgb = {'xgb__estimator__learning_rate': [0.1],
                 'xgb__estimator__max_depth': [15],
                 'xgb__estimator__n_estimators': [500]}
```

```python
xgb_clf = GridSearchCV(pipeline_xgb, param_grid = param_grid_xgb, scoring='precision_micro', cv=5, iid = False)
```

```python
xgb_clf.get_params().keys()
```

```python
xgb_clf.fit(X_train, y_train)
```

```python
xgb_clf_best = xgb_clf.best_estimator_
xgb_clf.best_score_
```

```python
xgb_clf.cv_results_['std_test_score']
```

```python
start = time.time()
cv_folds = 5
scores = cross_val_score(xgb_clf_best, X_train, y_train, cv=cv_folds, scoring='precision_micro')
print('Accuracy of each version: {}'.format(scores))
print('Mean accuracy: {}'.format(scores.mean()))
print('Standard error: {}'.format(scores.std() / cv_folds**0.5))
end = time.time()
print('Time: {}'.format(end - start))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

def evaluate(y,y_hat,labels):
  print(classification_report(y,y_hat))
  cm = confusion_matrix(y,y_hat)
  cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
  cmat = pd.DataFrame(cm)
  cmat.columns = labels
  cmat.set_index([pd.Index(labels, '')],inplace=True)
  sns.heatmap(cmat,cmap="YlGnBu", annot=True)
  plt.title("Confusion Matrix")
```

```python
y_pred = xgb_clf_best.predict(X_train)
evaluate(y_train, y_pred, labels)
accuracy_score(y_train,y_pred)
```

```python
f1_score(y_train, y_pred,average='micro')
```
